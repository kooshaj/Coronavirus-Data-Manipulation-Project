```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plots
import math
import ipywidgets as widgets
from IPython.display import display
from datetime import datetime
```

```python
covid1 = pd.read_csv("covid_data_1.csv")
covid1
```

```python
def death_rate_calculator(row):
    total_deaths = row["total_deaths"]
    pos_cases = row["total_cases"]
    if (np.isnan(pos_cases) or (np.isnan(total_deaths))):
        return None
    if ((np.isnan(pos_cases) == 0 )or (np.isnan(total_deaths)) == 0):
        return 0
    else:
        return total_deaths / pos_cases
def pos_rate_calculator(row):
    pos_cases = row["total_cases"]
    total_tests = row["total_tests"]
    if (np.isnan(pos_cases) or (np.isnan(total_tests))):
        return None
    if ((np.isnan(pos_cases) == 0 )or (np.isnan(total_tests)) == 0):
        return 0
    else:
        return pos_cases / total_tests
covid1["death_rate"] = covid1.apply(death_rate_calculator, axis = 1)
covid1["positive_rate"] = covid1.apply(pos_rate_calculator, axis = 1)
covid1 = covid1.loc[covid1.location != "International"]
covid1
```

```python
death_data = ["new_deaths", "new_deaths_per_million", "total_deaths", "total_deaths_per_million", "death_rate"] # add death rate(and cadiovascular deaths?)
cases_data = ["new_cases", "new_cases_per_million", "total_cases", "new_cases_per_million"]
test_data = ["new_tests", "total_tests", "total_tests_per_thousand", "new_tests_per_thousand", "tests_per_case", "positive_rate"] # tests_units(?)
population_data = ["population", "population_density", "median_age", "aged_65_older", "aged_70_older", "gdp_per_capita", "extreme_poverty"]
misc_data = ["diabetes_prevalence", "female_smokers", "male_smokers", "handwashing_facilities", "hospital_beds_per_thousand", "life_expectancy", "human_development_index", "stringency_index"]
```

```python
ALL = 'World'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.remove("World")
    unique.insert(0, ALL)
    return unique

Today = 'Today'
def unique_sorted_values_plus_today(array):
    unique = array.unique().tolist()
    unique.sort(reverse = True)
    unique.insert(0, Today)
    return unique
```

```python
large_output = widgets.Output()
output_value = widgets.Output()

dropdown_category = widgets.Dropdown(description = "Category", options = ['death', 'cases', 'test', "population", "misc", "correlations"])
dropdown_location = widgets.Dropdown(description = "Location", options = unique_sorted_values_plus_ALL(covid1.location))

filtered = pd.DataFrame()
filt_df = pd.DataFrame()

def common_filtering(category, locations):
    
    global filtered
    tempcovid = covid1.copy(deep = True)
    filtered = tempcovid
    cols = list(tempcovid)
    large_output.clear_output()
    templist = ['date', 'location']
    
    if locations != ALL:
        filtered = filtered[filtered.location == locations]
        if (category == 'death'):
            data = death_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'cases'):
            data = cases_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'test'):
            data = test_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'population'):
            data = population_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'misc'):
            data = misc_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
    else:
        if (category == 'death'):
            data = death_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'cases'):
            data = cases_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'test'):
            data = test_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'population'):
            data = population_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
        if (category == 'misc'):
            data = misc_data
            for i in data:
                templist.append(i)
            filtered = filtered[templist]
    
    with large_output:
        display(filtered)

def on_change1(category, locations):
    data1 = death_data
    output_value.clear_output()
    with output_value:
        if (category == 'death'):
            data1 = death_data
        if (category == 'cases'):
            data1 = cases_data
        if (category == 'test'):
            data1 = test_data
        if (category == 'population'):
            data1 = population_data
        if (category == 'misc'):
            data1 = misc_data
#         if (change.new == 'correlation'):
#             data1 = death_data
        d2 = widgets.Dropdown(description = category, options = data1)
        def on_change2(change1):
            global filt_df
            output_value.clear_output()
            with output_value:
                display(d2)
                filt_df = filtered.loc[filtered.location == locations]
                filt_df = filt_df[["date", "location", change1.new]]
                display(filt_df)
#                 display(filtered[["date", "location", change1.new]])
        d2.observe(on_change2, names= 'value')
        display(d2)

def dropdown_category_eventhandler(change):
    common_filtering(change.new, dropdown_location.value)
    on_change1(change.new, dropdown_location.value)
def dropdown_location_eventhandler(change):
    common_filtering(dropdown_category.value, change.new)
    on_change1(dropdown_category.value, change.new)
def subcat(change):
    on_change1(dropdown_category.value, dropdown_location.value)
    
    
    
dropdown_category.observe(subcat, names = 'value')
dropdown_category.observe(dropdown_category_eventhandler, names='value')
dropdown_location.observe(dropdown_location_eventhandler, names='value')

display(dropdown_category)
display(dropdown_location)
```
